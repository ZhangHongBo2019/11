### 双向循环链表
#### 相关操作：
##### 1.is _empty()链表是否为空
##### 2.length()链表的长度
##### 3.travel()遍历链表
##### 4.add()头结点插入
##### 5.append()尾结点插入
##### 6.insert()任意位置插入
##### 7.remove()删除结点
##### 8.search()定位结点

```python
class Node(object):
    """双向链表节点"""
    def __init__(self,elem):
        self.item = elem
        self.prev = None
        self.next = None
```

```python
class DoubleLinkList(object):
    """双链表类"""
    def __init__(self,node=None):
        self.__head = node
    
    """判断是否为空"""
    def is_empty(self):
        return self.__head is None
    
    """链表长度"""
    def length(self):
        cur = self.__head
        count = 0
        while cur!=None:
            count+=1
            cur = cur.next
        return count
    
    """遍历链表"""
    def trave(self):
        if self.is_empty():
            return 
        else:
            cur = self.__head
            while cur.next!=None:
                print(cur.item,end=" ")
                cur = cur.next
            print(cur.item,end=" ")
            
    """头部插入"""
    def add(self,elem):
        node = Node(elem)
        if self.is_empty():
            self.__head = node
        node.next = self.__head
        self.__head = node
        node.next.prev = node
        
    """尾部插入"""
    def append(self,elem):
        node = Node(elem)
        if self.is_empty():
            self.__head = node
        else:
            cur = self.__head
            while cur.next!=None:
                cur = cur.next
            cur.next = node
            node.prev = cur
            
    """指定位置插入"""
    def insert(self,pos,elem):
        if pos <=0:
            self.add(elem)
        elif pos >= (self.length()-1):
            self.append(elem)
        else:
            count = 1
            cur = self.__head
            node = Node(elem)
            while cur.next!=None:
                if count < pos-1:
                    count+=1
                    cur = cur.next
                else:
                    node.next = cur
                    node.prev = cur.prev
                    cur.prev.next = node
                    cur.prev = node
        
    """查找节点"""
    def search(self,elem):
        if self.is_empty():
            return False
        else:
            cur = self.__head
            count = 1
            while cur.next!=None:
                if cur.item!=elem:
                    cur = cur.next
                    count+=1
                else:
                    break
        return count
    
    """删除节点"""
    def remove(self,elem):
        if self.is_empty():
            return
        else:
            cur = self.__head
            while cur.next!=None:
                if cur.item!=elem:
                    cur = cur.next
                else:
                    cur.next.prev = cur.prev
                    cur.prev.next = cur.next
```
