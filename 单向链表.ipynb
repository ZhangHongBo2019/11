### 单向链表
#### 相关操作：
##### 1.is _empty()链表是否为空
##### 2.length()链表的长度
##### 3.travel()遍历链表
##### 4.add()头结点插入
##### 5.append()尾结点插入
##### 6.insert()任意位置插入
##### 7.remove()删除结点
##### 8.search()定位结点

```python
class Node(object):
    """节点"""
    def __init__(self,elem):
        self.elem = elem
        self.next = None
```

```python
class Single_Link_List(object):
    """头节点"""
    def __init__(self,node=None):
        self.__head = node
        node.next = node
        
    """判断链表是否为空"""
    def is_empty(self):
        return self.__head == None
    
    """链表的长度"""
    def length(self):
        if self.is_empty():
            return 0
        cur = self.__head
        count=1
        while cur.next!=None:
            cur = cur.next
            count+=1
        return count
    
    """遍历整个链表"""
    def trave(self):
        if self.is_empty():
            return
        cur = self.__head
        while cur.next!=None:
            print(cur.item,end=" ")
            cur = cur.next
        print(cur.item,end=" ")
        
    """头部插入"""
    def add(self,elem):
        node = Node(elem)
        if self.is_empty():
            self.__head = node
            node.next = None
        else:
            node.next = self.__head
            self.__head = node
    
    """尾部插入"""
    def append(self,elem):
        node = Node(elem)
        if self.is_empty():
            self.__head = node
        else:
            cur = self.__head
            while cur.next!=None:
                cur = cur.next
            cur.next = node
    
    """任意位置插入"""
    def insert(self,pos,elem):
        if pos <= 0:
            self.add(elem)
        elif pos >=(self.length()-1):
            self.append(elem)
        else:
            count = 0
            pre = self.__head
            while count < pos-1:
                count+=1
                pre = pre.next
            node = Node(elem)
            node.next = pre.next
            pre.next = node
    
    """定位节点的位置"""
    def search(self,elem):
        if self.is_empty():
            return False
        count = 1
        cur = self.__head
        while cur.next!=None:
            if cur.item!=elem:
                count+=1
                cur = cur.next
            else:
                break
        return count
    
    """删除单个节点"""
    def remove(self,elem):
        if self.is_empty():
            return False
        else:
            cur = self.__head
            pre = None
            while cur.next!=None:
                if cur.item!=elem:
                    pre = cur
                    cur = cur.next
                else:
                    pre.next = cur.next
```
